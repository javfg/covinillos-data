```python
from datetime import datetime
import json

import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

data_spain_path = "./spain_original_timeseries_2020-05-21.csv"
```

```python
data = pd.read_csv(data_spain_path).fillna(0)
```

```python
data['FECHA'] =  pd.to_datetime(data['FECHA'], format='%d/%m/%Y')
```

```python
data = data.groupby('FECHA').sum()
```

```python
data[data.columns] = data[data.columns].astype('int', errors='ignore')
```

```python
data['CASOSTOTALES'] = data['PCR+'] + data['TestAc+']
```
